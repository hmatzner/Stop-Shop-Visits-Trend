```python
# We group all the variables according to the 'device_id', showing also the number of visits ('count').
# We also show the rest of the variables' average.
df3 = df2.groupby('device_id').agg(count=('device_id', 'count'), 
                                   visit_time_spent=('visit_time_spent', 'mean'), 
                                   visit_weight=('visit_weight', 'mean'),
                                   home_distance_from_shop=('home_distance_from_shop', 'mean'))
df3.rename(columns={'visit_time_spent': 'avg_visit_time_spent', 
                    'visit_weight':'avg_visit_weight', 
                    'home_distance_from_shop': 'avg_home_distance_from_shop'}, inplace=True)
```

```python
# We calculate the total visit weight for each 'device_id'.
df3['total_visit_weight'] = df3['count'] * df3['avg_visit_weight']
df3
```

```python
# We express 'visit_time_spent' in minutes.
df3['avg_visit_time_spent'] = df3['avg_visit_time_spent'].dt.total_seconds().div(60)
```

```python
# We round the numbers.
df3['avg_visit_time_spent'] = df3['avg_visit_time_spent'].round(1)
df3['avg_visit_weight'] = df3['avg_visit_weight'].round(1)
df3['total_visit_weight'] = df3['total_visit_weight'].round(1)
df3['avg_home_distance_from_shop'] = df3['avg_home_distance_from_shop'].round(2)
```

```python
df3.columns
```

```python
# We reorder the columns...
df3 = df3[['count', 'avg_visit_time_spent', 'avg_home_distance_from_shop', 'avg_visit_weight', 'total_visit_weight']]
```

```python
# We can see there is no significant correlation in any variable, except for the ones that
# directly affect each other because of its definition, such as 'count' and 'total_visit_weight',
# given that 'total_visit_weight' = 'avg_visit_weight' * 'count'.
# So we can say that for the rest of them, there is no proof against their independency of one another.
df3.corr()
```

```python
# Given this fact, let's do some data normalization to define a loyal customer. 
# We want to know who are the special customers in which we want to put our marketing efforts, and we
# want that number to be much smaller than the 1928 devices that are considered active users.
# We have three parameters to determine if a customer is loyal:
# - 'total_visit_weight' (which is the result of the product between 'count' and 'avg_visit_weight'),
# - 'avg_visit_time_spent' and,
# - 'avg_home_distance_from_shop'.
# Each customer will get a grade (/100) according to this parameters.
# After an analysis of what best represents the loyalty of a customer, we determine that
# the impact of each one on them in the grade of each device will be respectively 20%, 15%, 65%.

x,y,z = 0.2,0.15,0.65

# According to FinancesOnline, 39% of customers are considered loyal, so we want to approximate that number.
# In order to do that, the 'loyal_customer_grade should be greater than 18 (/ 100), so around 30% of
# customers will be considered loyal.

# Reference: https://financesonline.com/customer-loyalty-statistics/#:~:text=73%25%20of%20companies%20depend%20more,Customer%20Communications%20Group%2C%202020
```

```python
# We do some data normalization with the 'min-max' method:
df4 = df3[['avg_visit_time_spent', 'avg_home_distance_from_shop', 'total_visit_weight']]
```

```python
# For the home distance, we do '1 minus'... so the closer the customer is to the shop, the greater its grade.
df4['avg_visit_time_spent'] = (df4['avg_visit_time_spent'] - df4['avg_visit_time_spent'].min()) / (df4['avg_visit_time_spent'].max() - df4['avg_visit_time_spent'].min())
df4['avg_home_distance_from_shop'] = 1 - (df4['avg_home_distance_from_shop'] - df4['avg_home_distance_from_shop'].min()) / (df4['avg_home_distance_from_shop'].max() - df4['avg_home_distance_from_shop'].min()) 
df4['total_visit_weight'] = (df4['total_visit_weight'] - df4['total_visit_weight'].min()) / (df4['total_visit_weight'].max() - df4['total_visit_weight'].min()) 
```

```python
# Now all values range from 0 (if they are the min of the column) to 1 (if they are the max of the column).
df4
```

```python
# We rename the columns accordingly...
df4.rename(columns={'avg_visit_time_spent':'norm_avg_visit_time_spent', 
                    'avg_home_distance_from_shop': 'norm_avg_home_distance_from_shop', 
                    'total_visit_weight': 'norm_total_visit_weight'}, inplace=True)
```

```python
# We now create the 'loyal_customer_grade' that will range from 0 to 100.
df4['loyal_customer_grade'] = ((df4['norm_avg_visit_time_spent'] * x) + (df4['norm_avg_home_distance_from_shop'] * y) + (df4['norm_total_visit_weight'] * z)) * 100
```

```python
df4
```

```python
# And now we create the dataframe 'df_loyal' that shows only the loyal customers.
df_loyal = df4[df4['loyal_customer_grade'] > 18]
```

```python
# And we also create a dataframe that shoes only the casual customers.
df_casual = df4[df4['loyal_customer_grade'] <= 18]
```
